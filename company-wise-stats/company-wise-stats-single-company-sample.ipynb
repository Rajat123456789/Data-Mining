```python
output = response.json()['pageProps']['company']['metrics']
```

```python
import csv
import json

data = output
file = open('output.csv', 'w', newline='')
csvwriter = csv.writer(file)
count = 0
for i in data:
    if count == 0:
        header = i.keys()
        csvwriter.writerow(header)
        count += 1
    csvwriter.writerow(i.values())
file.close()
print("output.csv created successfully")
```
