```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

companies_data = pd.read_csv('simplify-company-data.csv')
```

number of companies founded in each year

```python
plt.figure(figsize=(20, 10))
companies_data['year_founded'].value_counts().sort_index(ascending=False).plot(kind='bar')

```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='year_founded', hue='customer_type', data=companies_data)
plt.xticks(rotation=90)
plt.xlabel('Year Founded', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Count of Companies by Year Founded and Customer Type', fontsize=16)
plt.show()
```

```python
companies_data.groupby(['year_founded', 'customer_type']).size().unstack().plot(kind='bar', stacked=True, figsize=(20, 10))
```

```python
# Replace NaN values in 'year_founded' column with 0
companies_data['year_founded'].fillna(0, inplace=True)

# Convert 'year_founded' to integer
companies_data['year_founded'] = companies_data['year_founded'].astype(int)

# Plot
plt.figure(figsize=(20, 20))
companies_data.groupby(['year_founded', 'customer_type']).size().unstack().plot(kind='bar', stacked=True)
plt.xticks(rotation=45)  # To keep the years horizontal
plt.xlabel('Year Founded', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Distribution of Customer Types Over the Years', fontsize=16)
plt.legend(title='Customer Type', fontsize=12)
plt.show()
```

```python

import matplotlib.pyplot as plt

# Assuming you have the data in a DataFrame called 'companies_data'
grouped = companies_data.groupby(['year_founded', 'customer_type']).size().unstack()

# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(20, 10))

# Plot the stacked bar chart
grouped.plot(kind='bar', stacked=True, ax=ax)

# Set the x-axis tick labels to the year_founded values
ax.set_xticklabels(grouped.index, rotation=0)

# Add axis labels and title
ax.set_xlabel('Year Founded')
ax.set_ylabel('Count')
ax.set_title('Companies by Year Founded and Customer Type')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(20, 10))
companies_data.groupby('funding_stage')['funding_total'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=75)
plt.xlabel('Funding Stage', fontsize=14)
plt.ylabel('Mean Funding Total', fontsize=14)
plt.title('Mean Funding Total by Funding Stage', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10_companies = companies_data['funding_stage'].value_counts().nlargest(10).index.tolist()
subset_data = companies_data[companies_data['funding_stage'].isin(top_10_companies)]
subset_data.groupby('funding_stage')['funding_total'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=75)
plt.xlabel('Funding Stage', fontsize=12)
plt.ylabel('Mean Funding Total', fontsize=12)
plt.title('Mean Funding Total by Funding Stage (Top 10 Companies)', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
companies_data.groupby('funding_stage')['funding_total'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=75)
plt.xlabel('Funding Stage', fontsize=14)
plt.ylabel('Mean Funding Total', fontsize=14)
plt.title('Mean Funding Total by Funding Stage', fontsize=16)
plt.show()
```

```python
#the average amount of funding raised by companies in each funding stage y axis should be the average amount of funding raised x axis should be the funding stage in descending order of the average amount of funding raised

plt.figure(figsize=(20, 10))
companies_data.groupby('funding_stage')['funding_total'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=75)
plt.xlabel('Funding Stage', fontsize=14)
plt.ylabel('Mean Funding Total', fontsize=14)
plt.title('Mean Funding Total by Funding Stage', fontsize=16)
plt.show()
```
